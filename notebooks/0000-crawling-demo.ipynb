```python
import requests
from time import sleep
from modules.util import jsonify, time_gap, load_json
from modules.crawl import reviewAPI
```

```python
HAS_NEXT_PAGE = False
NEXT_CURSOR = ""
```

```python
crawled_review = reviewAPI().get(
    uuid="0f101f8c-ec09-39c4-9be0-2f9cc464d332", cursor=NEXT_CURSOR, page_count=20
).text
crawled_review = jsonify(crawled_review)
HAS_NEXT_PAGE = crawled_review["pageInfo"]["hasNextPage"]
HAS_PREV_PAGE = crawled_review["pageInfo"]["hasPreviousPage"]
NEXT_CURSOR = (
    crawled_review["pageInfo"]["endCursor"]
    if ("endCursor" in crawled_review["pageInfo"].keys()) & HAS_NEXT_PAGE
    else ""
)
# HAS_NEXT_PAGE, HAS_PREV_PAGE, crawled_review
```

```python
NEXT_CURSOR = ""
while HAS_NEXT_PAGE:
    next_reviews = (
        reviewAPI().get(uuid="0f101f8c-ec09-39c4-9be0-2f9cc464d332", cursor=NEXT_CURSOR, page_count=20).text
    )
    next_reviews = jsonify(next_reviews)
    
    # Save
    crawled_review["reviews"].extend(next_reviews["reviews"])
    
    HAS_NEXT_PAGE = next_reviews["pageInfo"]["hasNextPage"]
    HAS_PREV_PAGE = next_reviews["pageInfo"]["hasPreviousPage"]
    NEXT_CURSOR = (
        next_reviews["pageInfo"]["endCursor"]
        if ("endCursor" in next_reviews["pageInfo"].keys()) & HAS_NEXT_PAGE
        else ""
    )
    
    if not HAS_NEXT_PAGE:
        break
    sleep(time_gap(mu=4))
```
